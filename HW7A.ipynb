### Problem 1: user input

* Prompt the user to enter their first name and store it in a variable called "first".
* Prompt the user to enter their last name and store it in a variable called "last".
* Prompt the user to enter their percentage in ChemE 375 and store it in a variable called "pct".
* Convert the percentage to a letter grade called "grade".
* Then print the person's name, and grade; for example:
    "Tim Smith: A"


```python
import numpy as np
import matplotlib.pyplot as plt

first = input("First Name: ")
last = input("Last Name: ")
pct = input("Enter Score: ")
pct = float(pct)

def grade(pct):
        if pct >93 and pct< 100:
            return "A"
        if pct < 93 and pct > 90:
            return 'A-'
        if pct < 90 and pct> 87:
            return 'B+'
        if pct < 87 and pct> 83:
            return 'B'
        if pct < 83 and pct> 80:
            return 'B-'
        if pct < 80 and pct> 77:
            return 'C+'
        if pct < 77 and pct> 73:
            return 'C'
        if pct < 73 and pct> 70:
            return 'C-'
        if pct < 70 and pct> 67:
            return 'D+'
        if pct < 67 and pct> 60:
            return 'D'
        return 'F'
    
print(first, last, ':')
print(grade(pct))
```

### Problem 2: Numpy file I/O

#### Part a
Given the following code, 
* Collect the data x,y,z,w into a single array called data.
* Write a plain numpy text data file called "data_1.dat". 

```python
x = np.linspace(0,2.0*np.pi,100)
y = np.sin(x)+np.random.randn(100)*0.2
z = np.cos(x)+np.random.rand(100)*0.1
w = x**0.2 + np.random.rand(100)*0.3
plt.plot(x,y,x,z,x,w)

#---- SAVE THE DATA TO AN ARRAY AND WRITE IT TO data_1.dat
data = np.vstack([x,y,z,w]).T
print(data[0:5,:])

x = x.reshape(-1,1)
y = y.reshape(-1,1)
z = z.reshape(-1,1)
w = w.reshape(-1,1)
data = np.concatenate((x,y,z,w),axis=1)
print(data[0:5,:])

np.savetxt('data_1.txt', data)
```

#### Part c

```python
newData = np.loadtxt('data_2.dat',delimiter=',')
x = newData[:,0]
y = newData[:,1]
w = newData[:,2]
z = newData[:,3]
plt.plot(x,y,label='y(m/s)')
plt.plot(x,z,label='z(K)')
plt.plot(x,w,label='w(m^3/s)')
plt.legend()
plt.xlabel('t(s)')
plt.ylabel('y,z,w')
```
